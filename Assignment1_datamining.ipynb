Explanation:

1. Given enough data, the central limit theorem (CLT) states that a given variable's distribution resembles a normal distribution.

2. This theorem is a key idea in statistics and is applied in many statistical tests. Because they offer a technique to represent data that might not follow a normal distribution, CLTs are significant.

3. For data that is not normally distributed, you can also employ techniques based on normal distributions (such z-scores). The law of large numbers leads to the central limit theorem.

4. As the sample size grows, the law of big numbers states that The mean is closer to the total of the sample. In addition to the sample value roughly resembling the population value, the central limit theorem goes one step farther and Nevertheless, the sample mean's distribution resembles a normal distribution.

5. The distribution of sample means for various sample sizes is displayed in the above histogram. The sample mean distribution resembles a normal distribution as sample size rises.

6. This is due to the central limit theorem.

## Question 2 (10 points) 

The exercise refers to Example 6.6 of the  [Math for ML book](https://mml-book.github.io/book/mml-book.pdf). 

Simulate and plot the bivariate normal distribution with the shown parameters.  
You need to use the [Cholesky factorization](https://numpy.org/doc/stable/reference/generated/numpy.linalg.cholesky.html) for the simulation. 

````python
import numpy as np
from scipy.linalg import cholesky
obs_number = 10000
x = [1,2,3]
sds = [1,2,3]
obser = np.vstack([np.random.normal(loc=mean, scale=sd, size=obs_number) 
                  for mean, sd in zip(x, sds)])
cor_matrix = np.array([[1.0,0.6,0.9], [0.6,1.0,0.5],[0.9,0.5,1.0]])
L = cholesky(cor_matrix, lower=True)
print(np.corrcoef(L.dot(obser)))
````
